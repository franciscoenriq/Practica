```python
#eliminamos los valores que contienen a los usuarios más activos que justamente no son estudiantes. 
staff_users = ['ayanez','uptime_check_user','admin','vdaiep','root','hmontenegro','drojas','fguinez','fredondo','cachorro','tpeet','mpeet','smartinez','csaintpierre','scabezas','testlogin',
               'usuarioprueba','drojastest','brodriguez','demo1','redfidtest2','demo2','redfiduserprueba','testuser3','testvdaiep','testuser2','aldoramirez','dhiguera','redfid']

df = df[~df['username'].isin(staff_users)]
df.username.value_counts()
```

```python
#tenemos un formato iso8601 para el tiempo  
#agregamos las columnas de semana y año para los registros. 
df = df.dropna(subset=['time'])
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df['semana'] = df['time'].dt.isocalendar().week
df['año'] = df['time'].dt.isocalendar().year
df
```

```python
#filtramos por usaurio 
df_filtrado = df[df['username'] == 'mfajardo']
#numero de filas y columnas 
df_filtrado.shape
#df_ordenado por tiempo 
df_ordenado = df_filtrado.sort_values(by='time')

#extraemos valores de tiempo para ver como trabajar con ellos
valor_tiempo1 = df_ordenado.iloc[0]['time']
valor_tiempo2 = df_ordenado.iloc[3]['time']

#en el arreglo tiempos queremos asociar cada evento al momento en que pasaron para cada usuario, en este caso solo es para un solo usuario. 
tiempos = []
for momento, evento in zip(df_ordenado.iloc[:]['time'], df_ordenado.iloc[:]['event_type']): 
    tupla = (momento, evento)
    tiempos.append(tupla)

#vemos que podemos transformar estos tiempos para poder manejarlos. 
tiempo1 = datetime.fromisoformat(valor_tiempo1)
tiempo2 = datetime.fromisoformat(valor_tiempo2)
print(tiempo1)
print(tiempo2-tiempo1)
print(tiempos)
print(len(tiempos))
```
